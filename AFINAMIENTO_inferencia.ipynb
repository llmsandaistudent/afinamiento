```python
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load tokenizer from the original model
tokenizer = AutoTokenizer.from_pretrained('distilgpt2')
GPT2model = AutoModelForCausalLM.from_pretrained('distilgpt2')
finetuned_model= model = AutoModelForCausalLM.from_pretrained('/content/drive/MyDrive/FINETUNING/resultados/checkpoint-400')
```

```python
#PROMPT PRE-TRAINED MODEL distiledGPT2

# Define the prompt
prompt = "Q: Explain intertia\nA:"

# Tokenize the prompt
inputs = tokenizer(prompt, return_tensors="pt")

# Generate text
output = GPT2model.generate(
    **inputs,
    max_length=100,  # longitud máxima del texto a generar
    num_return_sequences=1,  # Número de secuencias a generar
    no_repeat_ngram_size=2,  # Prevenir repetición de frases
    temperature=0.7,  # Control de aleatoriedad
    top_p=0.9,
    top_k=50
)

# Decode the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```

```python
#PROMPT FINE-TUNED MODEL

# Define the prompt
prompt = "Q: Explain intertia\nA:"

# Tokenize the prompt
inputs = tokenizer(prompt, return_tensors="pt")

# Generate text
output = finetuned_model.generate(
    **inputs,
    max_length=100,  # Set the max length for the generated text
    num_return_sequences=1,  # Number of sequences to generate
    no_repeat_ngram_size=2,  # Prevent repetition of phrases
    temperature=0.7,  # Control the randomness of the text
    top_p=0.9,  # Probability threshold for nucleus sampling
    top_k=50  # Limit the vocabulary
)

# Decode the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print(generated_text)
```
